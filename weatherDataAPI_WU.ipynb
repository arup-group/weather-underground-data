```python
import requests
from pprint import pprint
import json

key="fe9cfb7dc567447b9cfb7dc567147b93" #From buying a personal weather station
#stationID= "KCADALYC40" #Daly city(Arup station)
stationID= "KTXKATY32"
date = "20221104"

#past data
url = "https://api.weather.com/v2/pws/history/all?stationId="+stationID+"&format=json&units=m&date="+date+"&apiKey="+key

#current conditions
#url= "https://api.weather.com/v2/pws/observations/current?stationId="+stationID+"&format=json&units=e&apiKey="+key

print(url)

r = requests.get(url)
r.json()

data = r.json()
pprint(data)
```

```python
filename = 'test.txt'
with open(filename, 'w') as outfile:  
    json.dump(data, outfile)
```

```python
counter=0
listTemp =[]
listsolar=[]
listuv=[]
listRH=[]
listWD=[]
listWS=[]
listDateTime=[]
listWGAvg=[]
listWGHigh=[]
listWGLow=[]


for value in data['observations']:
    temp = data['observations'][counter]['metric']['tempAvg']
    solar = data['observations'][counter]['solarRadiationHigh']
    uv = data['observations'][counter]['uvHigh']
    hum= data['observations'][counter]['humidityAvg']
    wd= data['observations'][counter]['winddirAvg']
    ws= data['observations'][counter]['metric']['windspeedAvg']
    dateTime= data['observations'][counter]['obsTimeLocal']
    wgAvg= data['observations'][counter]['metric']['windgustAvg']
    wgHigh= data['observations'][counter]['metric']['windgustHigh']
    wgLow= data['observations'][counter]['metric']['windgustLow']
    
    counter=counter+1
    listTemp.append(temp)
    listsolar.append(solar)
    listuv.append(uv)
    listRH.append(hum)
    listWD.append(wd)
    listWS.append(ws)
    listDateTime.append(dateTime)
    listWGAvg.append(wgAvg)
    listWGHigh.append(wgHigh)
    listWGLow.append(wgLow)
    
print(listDateTime)
```

```python
from datetime import datetime

datetime_object = [datetime.strptime(x, '%Y-%m-%d %H:%M:%S') for x in listDateTime]

#for x in datetime_object:
 #   print(type(x))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(18,5))

ax.plot(datetime_object,listTemp, linewidth=0.5,marker='o', linestyle='-', label='Temp')
ax.set_xlabel('Time')
ax.set_ylabel('Temp')
ax.set_title('Temp');

#plt.savefig('Temp.png')
```
